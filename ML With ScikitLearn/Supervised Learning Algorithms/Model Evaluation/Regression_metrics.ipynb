### Regression Metrics

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error
import numpy as np
```

```python
RealMarks = [90 , 60 , 80 , 100] #Actual 

ModelPrediction = [85 , 70 , 70 , 95] #prediction 
```

```python
mae = mean_absolute_error(RealMarks , ModelPrediction)

mse = mean_squared_error(RealMarks , ModelPrediction) 

rmse = np.sqrt(mse)

print(f"MAE= {mae}")
print(f"MSE= {mse}")
print(f"RMSE= {rmse}")
```
