# Student Score Prediction 

```python
# import libraries 
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error , mean_squared_error
import numpy as np
```

```python
# create data 
student_data = {
    'Hours':[1 , 2 , 3 , 4 , 5 , 6],
    'Score':[40 , 50 , 62 , 75 , 88 , 97]
}
# convert dataframe 
df = pd.DataFrame(student_data)
df
```

```python
# df convert to csv file 
df.to_csv('StudentExamScore.csv')
```

```python
# now call csv 
my_df = pd.read_csv('C:/Users/Muhammad Yahya/Downloads/python/ML With ScikitLearn/Supervised Learning Algorithms/Projects/Project1_Use_Algo_Linear_Regression/StudentExamScore.csv' , index_col=[0])
my_df
```

```python
# new var 
X = my_df[['Hours']] #independent var / input features
y = my_df['Score'] #dependent var / label 
```

```python
# model 
model = LinearRegression() 
model = model.fit(X , y)  
```

```python
# prediction 
predicted_score = model.predict(X)
predicted_score
```

```python
# MODEL EVALUATION 
mae = mean_absolute_error(y , predicted_score)
mse = mean_squared_error(y , predicted_score)
rmse = np.sqrt(mse)
```

```python
print(f'MAE = {mae}')
print(f'MSE = {mse}')
print(f'RMSE = {rmse}')
```
