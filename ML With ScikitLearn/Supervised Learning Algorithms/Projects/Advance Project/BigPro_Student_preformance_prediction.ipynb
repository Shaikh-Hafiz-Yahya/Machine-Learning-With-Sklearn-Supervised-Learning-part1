# Big Project Student Performance Prediction

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error , mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Student_Performance.csv')
df.head(10)
```

```python
df.info()
```

```python
X = df[['study_hours' , 'attendance_percentage']]
y = df['overall_score']
```

```python
model = LinearRegression()
model = model.fit(X , y)
```

```python
model_predict = model.predict(X)
model_predict
```

```python
# model evaluation 
mae = mean_absolute_error(y , model_predict)
mse = mean_squared_error(y , model_predict)
rmse = np.sqrt(mse)
```

```python
print(f'MAE = {mae}')
print(f'MSE = {mse}')
print(f'RMSE = {rmse}')
```

```python
model_predict_unknown_input = model.predict([[9.2 , 9.88]])[0]
model_predict_unknown_input
```

```python
plt.plot(model_predict)
plt.xlabel('independent_features' , color='green')
plt.ylabel('target / label' , color='green')
plt.title('Student_Score_Prediction' , color='r')
plt.grid()
plt.show()
```

```python
plt.hist(model_predict , bins=10 , color='blue')
plt.xlabel('independent_features' , color='green')
plt.ylabel('target / label' , color='green')
plt.title('Student_Score_Prediction' , color='r')
plt.grid()
plt.show()
```
