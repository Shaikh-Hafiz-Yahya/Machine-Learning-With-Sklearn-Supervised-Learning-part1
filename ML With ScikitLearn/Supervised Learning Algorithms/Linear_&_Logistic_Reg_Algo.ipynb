## Simple Linear Regression Algo  

### Regression Problem

### Exam Marks Prediction Through Study Hrs

### Always; Numerical Output/Label 

```python
from sklearn.linear_model import LinearRegression

X = [[1] , [2] , [3] , [4] , [5] , [5.5]] #indep features
y = [40 , 50 , 65 , 80 , 90 , 94.5]#dep features

model = LinearRegression()
model = model.fit(X , y)

predicted_marks = model.predict([[1],[2],[4],[5],[1/2],[8.5]])
y_predict = predicted_marks
y_predict
```

## Logistic Regression Algo

### Classification Problem

### Exam Marks Prediction Through Study Hrs

```python
from sklearn.linear_model import LogisticRegression

X = [[1] , [2] , [3.5] , [4] , [4.5]] #(study hours) independent var/input data
y = [0 , 0 , 1 , 1 , 1] #(Pass = 1,Fail = 0) label/dependent var/output data

model = LogisticRegression()
model = model.fit(X , y)

result_predict = model.predict([[1/2],[1],[1.5],[3.5],[6],[7]])
result_predict
```

## Mini Project

### Weight Loss Prediction Through (Calories Burn).

### Binary Classification Problem

### Numerical Output/Label

```python
from sklearn.linear_model import LogisticRegression

X = [[0.0] , [0.25] , [0.47] , [0.5] , [0.99] , [1.99] , [2.98]]
y = [0 , 0 , 0 , 1 , 1 , 1 , 1]

model = LogisticRegression().fit(X,y)
weight_loss = model.predict([[2] , [0.5] , [2] , [79]])
weight_loss
```

### Weight Loss Prediction Through (Calories Burn).

### Binary Classification Problem
### Categorical Output/Label

```python
from sklearn.linear_model import LogisticRegression

X = [[0.0] , [0.25] , [0.47] , [0.5] , [0.99] , [1.99] , [2.98]]
y = ['No' , 'No' , 'No' , 'WeightLoss' , 'WeightLoss' , 'WeightLoss' , 'WeightLoss'] #WeightLoss=(1 or Yes or True) , No=(0 or No , False) 

model = LogisticRegression().fit(X,y)
weight_loss = model.predict([[2] , [0.5] , [2] , [79]])
weight_loss
```

### Cholesterol Disease Prediction

```python
from sklearn.linear_model import LogisticRegression

X = [[130 , 159]  , [40 , 50]]
y = ['Bad' , 'GoodForMale , GoodForFemale']

model = LogisticRegression()
model = model.fit(X , y)

check_cholestrol = model.predict([[130, 150] , [40 , 50]])

if check_cholestrol[0] == 'Bad':
    print('Border Line')
else:
    print('Good Chol')
```
