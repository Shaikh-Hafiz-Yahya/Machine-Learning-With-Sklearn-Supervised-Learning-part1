## Male & Female Features Prediction

```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
```

```python
# create Data 
gender_features = {
    'Hair Length(inch)':[3.1 , 1 , 2.5 , 6.99 , 11.5 , 17 , 15 , 22 , 4.5],
    'Hijab':[0 , 0 , 0 , 0 , 1 , 1 , 1 , 1 , 0],
    'Height(ft)':[5.01 , 6.2 , 4.1 , 5.99 , 4.99 , 6.01 , 4.1 , 5.7 , 5.8],
    'Gender':['Male' , 'Male' , 'Male' , 'Male' , 'Female' , 'Female' , 'Female' , 'Female' ,'Male']
    } 
df = pd.DataFrame(gender_features)
df
```

```python
# DATA FRAME CONVERT TO .CSV FILE
df_convert_csv = df.to_csv('Gender_Selection.csv')
```

```python
call_csv = pd.read_csv('C:/Users/Muhammad Yahya/Downloads/python/ML With ScikitLearn/Supervised Learning Algorithms/Gender_Selection.csv' , index_col=[0])
my_df = call_csv
my_df
```

```python
# create 2 variables
X = my_df[['Hair Length(inch)' , 'Hijab' , 'Height(ft)']] #independent var / features
y = my_df['Gender'] #target var / label
```

```python
# build & fit model 
model = DecisionTreeClassifier()
model = model.fit(X , y)
```

```python
actual = pd.DataFrame([[3.10 , 0 , 5.01]] , columns=['Hair Length(inch)' , 'Hijab' , 'Height(ft)'])
actual = model.predict(actual)[0]
print(actual)

predicted = pd.DataFrame([[19 , 0 , 6]] , columns=['Hair Length(inch)' , 'Hijab' , 'Height(ft)'])
predicted = model.predict(predicted)[0]
print(predicted)
```

```python
if (actual == 'Male' and predicted == 'Male'):
    print('You are allowed "male" class')
else:    
    print('You are allowed "female" class')
```
