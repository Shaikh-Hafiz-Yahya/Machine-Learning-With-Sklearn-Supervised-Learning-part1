```python
import pandas as pd
```

```python
dic_data = {'Name':['Ali' , 'Umer' , 'Aatif' , 'Ayan'],
            'Age':[22 , None , 31 , None],
            'Salary':[120000 , 100000 , None , None]}
df = pd.DataFrame(dic_data)
print('Original DataFrame')
df 
```

```python
df.isnull().mean()
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
# fillna Age k column main null values ko fill krein
df['Age'].fillna(df['Age'].mean() , inplace=True)
df
```

```python
# fillna Salary k column main null values ko fill krein
df['Salary'].fillna(df['Salary'].mean() , inplace=True)
df
```

```python
data = {
    "Name": ["Ali", "Sara", "Ahmed", "Ayesha"],
    "Gender": ["Male", "Female", "Male", "Female"],
    "City": ["Lahore", "Karachi", "Lahore", "Islamabad"]
}

df = pd.DataFrame(data)
print(df)
```

```python
df_encoded = pd.get_dummies(df , columns=['City'])
df_encoded
```
